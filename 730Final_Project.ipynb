```python
import pandas as pd
import numpy as np
import re
```

```python
all_participants_df = pd.read_csv("participants.csv")
rooms_df = pd.read_csv("rooms.csv")
```

```python
all_participants_df
```

```python
team0 = all_participants_df[all_participants_df['Team'] == 'Team0'].reset_index()['index'].values.tolist()
team1 = all_participants_df[all_participants_df['Team'] == 'Team1'].reset_index()['index'].values.tolist()
team2 = all_participants_df[all_participants_df['Team'] == 'Team2'].reset_index()['index'].values.tolist()
team3 = all_participants_df[all_participants_df['Team'] == 'Team3'].reset_index()['index'].values.tolist()
team4 = all_participants_df[all_participants_df['Team'] == 'Team4'].reset_index()['index'].values.tolist()
team5 = all_participants_df[all_participants_df['Team'] == 'Team5'].reset_index()['index'].values.tolist()
team6 = all_participants_df[all_participants_df['Team'] == 'Team6'].reset_index()['index'].values.tolist()
team7 = all_participants_df[all_participants_df['Team'] == 'Team7'].reset_index()['index'].values.tolist()
team8 = all_participants_df[all_participants_df['Team'] == 'Team8'].reset_index()['index'].values.tolist()
team9 = all_participants_df[all_participants_df['Team'] == 'Team9'].reset_index()['index'].values.tolist()
```

```python
room_limit = rooms_df.iloc[:,1].values.tolist()
```

```python
dummies1 = pd.get_dummies(all_participants_df[['Priority1','Priority2']])
dummies2 = pd.get_dummies(all_participants_df[['Priority1','Priority2']])
obj_coef = pd.concat([dummies1,dummies2],axis = 'columns')
obj_coef = obj_coef.values.tolist()
```

```python
num_activity = 5
num_room = 5
num_person = len(all_participants_df)
num_activityAMPM = 10
```

```python
#Model
model = ConcreteModel()

#DVs
model.x1 = Var(range(num_room), range(num_activity), domain = Binary)
model.x2 = Var(range(num_person),range(num_activityAMPM), domain = Binary)

#Objective
model.Objective = Objective(expr = sum(obj_coef[i][j]*model.x2[i,j] for i in range(num_person) for j in range(num_activityAMPM)), sense = maximize)

```

```python
#Group Constraint
model.groupconstraints = ConstraintList()
for i in team0:
  for j in range(num_activityAMPM):
    model.groupconstraints.add(expr = model.x2[i,j] - model.x2[34,j] == 0)
for i in team1:
  for j in range(num_activityAMPM):
    model.groupconstraints.add(expr = model.x2[i,j] - model.x2[0,j] == 0)
for i in team2:
  for j in range(num_activityAMPM):
    model.groupconstraints.add(expr = model.x2[i,j] - model.x2[9,j] == 0)
for i in team3:
  for j in range(num_activityAMPM):
    model.groupconstraints.add(expr = model.x2[i,j] - model.x2[12,j] == 0)
for i in team4:
  for j in range(num_activityAMPM):
    model.groupconstraints.add(expr = model.x2[i,j] - model.x2[15,j] == 0)
for i in team5:
  for j in range(num_activityAMPM):
    model.groupconstraints.add(expr = model.x2[i,j] - model.x2[17,j] == 0)
for i in team6:
  for j in range(num_activityAMPM):
    model.groupconstraints.add(expr = model.x2[i,j] - model.x2[21,j] == 0)
for i in team7:
  for j in range(num_activityAMPM):
    model.groupconstraints.add(expr = model.x2[i,j] - model.x2[26,j] == 0)
for i in team8:
  for j in range(num_activityAMPM):
    model.groupconstraints.add(expr = model.x2[i,j] - model.x2[28,j] == 0)
for i in team9:
  for j in range(num_activityAMPM):
    model.groupconstraints.add(expr = model.x2[i,j] - model.x2[33,j] == 0)
```

```python
#RoomAssign_ToActivity Contraint
model.room_activity_constraints = ConstraintList()
for i in range(num_room):
  model.room_activity_constraints.add(expr = sum(model.x1[i,j] for j in range(num_activity)) == 1)
for j in range(num_activity):
  model.room_activity_constraints.add(expr = sum(model.x1[i,j] for i in range(num_room)) == 1)

```

```python
#Number of people room can contain Constraint
model.capacityconstraints = ConstraintList()
for j in range(num_activity):
  model.capacityconstraints.add(expr = sum(room_limit[i]*model.x1[i,j] for i in range(num_room))-sum(model.x2[i,j] for i in range(num_person)) >= 0)
for j in range(num_activity):
  model.capacityconstraints.add(expr = sum(room_limit[i]*model.x1[i,j] for i in range(num_room))-sum(model.x2[i,j+5] for i in range(num_person)) >= 0)
```

```python
#Everyperson should attend two different activities in AM and PM Constraint
model.people_constraints = ConstraintList()
for i in range(num_person):
  model.people_constraints.add(expr = sum(model.x2[i,j] for j in range(num_activity)) == 1)
for i in range(num_person):
  model.people_constraints.add(expr = sum(model.x2[i,j+5] for j in range(num_activity)) == 1)
for i in range(num_person):
  model.people_constraints.add(expr = model.x2[i,0] + model.x2[i,5] <= 1)
  model.people_constraints.add(expr = model.x2[i,1] + model.x2[i,6] <= 1)
  model.people_constraints.add(expr = model.x2[i,2] + model.x2[i,7] <= 1)
  model.people_constraints.add(expr = model.x2[i,3] + model.x2[i,8] <= 1)
  model.people_constraints.add(expr = model.x2[i,4] + model.x2[i,9] <= 1)
```

```python
opt = SolverFactory('glpk')
results = opt.solve(model, tee=True)
```
